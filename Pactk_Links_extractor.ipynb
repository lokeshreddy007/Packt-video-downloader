```
# Initize the required library
import requests
from bs4 import BeautifulSoup
```

```
# Add Url Of the Course
baseurl = "https://www.packtpub.com/product/css-the-complete-guide-incl-flexbox-grid-and-sass-video/9781789954692"
html = requests.get(baseurl).text
soup = BeautifulSoup(html,'html.parser')
print(soup)
```

```
# Get All the Links and URL of this Course
books = []
links = []
for bookname in soup.findAll('a', attrs={'class':'links__item toc-index'}):
    books.append(bookname.text)
for link in soup.findAll('a', attrs={'class':'links__item toc-index'}):
    links.append(link.get('href','/n''/n'))
```

```
print(books)
```

```
print(links)
```
